```python
import random

def play_game_RPS():
    user_score = 0
    computer_score = 0
    choices = ["rock", "paper", "scissors"]
    want_to_play = input("\nWant to play the game? (y/n)").lower()
    if want_to_play == "y":
            print("\n  Welcome to Rock-Paper-Scissors! ")
        
            print("\n  Instructions: Type 'rock', 'paper', or 'scissors' to play.")
        
            print("\n  Type 'exit' at any time to end the game.")
    else:
        print("Next Time")
        return

    while True:
        user_choice = input("\nYour choice: ").lower()

        if user_choice == 'exit':
            break
        
        if user_choice not in choices:
            print("\nInvalid choice. Please try again.")
            continue

        computer_choice = random.choice(choices)
        print(f"\n'Computer chose: {computer_choice}'")

        if user_choice == computer_choice:
            print("\nIt's a tie!")
        elif (user_choice == "rock" and computer_choice == "scissors") or \
             (user_choice == "scissors" and computer_choice == "paper") or \
             (user_choice == "paper" and computer_choice == "rock"):
            print("\nYou win this round!")
            user_score += 1
        else:
            print("\nYou lose this round!")
            computer_score += 1

        print(f"\nScore -> You: {user_score} | Computer: {computer_score}")

        play_again = input("\nPlay another round? (y/n): ").lower()
        if play_again == 'y':
            
            print("\nWelcome to Rock-Paper-Scissors!")
            print("\nInstructions: Type 'rock', 'paper', or 'scissors' to play.")
            print("\nType 'exit' at any time to end the game.")
            continue 
        else:
            break

    print("\nFinal Game Summary")
    print(f"\nFinal Score is You: {user_score} | Computer: {computer_score}")
    print("\nThanks for playing!")

if __name__ == "__main__":
    play_game_RPS()
```
