```python
your_number1 = float(input("Tell me First number of choice:"))
your_number2 = float(input("Tell me second number of choice:"))

print("Tell me your Prefences from (1/2/3/4/5):")
print("1. Addition")
print("2. Substraction")
print("3. Multiplication")
print("4. Division")
print("5. Reminder")
while True:
    choice = input("what you want to be calcauted:")
    if choice not in  ("1","2","3","4"):
        print("Enter from the above Prefences only '1/2/3/4/5'")
        continue

    if choice == "1":
         result = your_number1 + your_number2
         print(f" The result is {result}")
    elif choice =="2":
        result = your_number1 - your_number2
        print(f" The result is {result}")
    elif choice == "3":
        result = your_number1 * your_number2
        print(f" The result is {result}")
    elif choice =="4":
        if your_number2== 0: 
            print("Division by zero is not supported")
        else:    
            result = your_number1/your_number2
            print(f" The result is {result}")
    elif choice == "5":
        if your_number2== 0: 
            print("Reminder by zero is not supported")
        else:    
            result = your_number1%your_number2
            print(f" The result is {result}")
    else:
        print("Invalid! Read the instruction carefully")
    break

    
```
