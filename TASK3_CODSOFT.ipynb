```python
import random
import string

def custom_password():
    print("Professional Password Generator:")
    
    try:
        length = int(input("Enter password length: "))
        
        print("\nSelect Complexity Level:")
        print("1. Easy (Letters (Lowercase  + Uppercase)")
        print("2. Medium (Letters(Lowercase + Uppercase) + Numbers)")
        print("3. Hard (Letters (Uppercase + Lowercase) + Numbers + Symbols)")
        choice = input("Enter choice (1/2/3): ")

        lower = string.ascii_lowercase
        upper = string.ascii_uppercase
        digits = string.digits
        symbols = string.punctuation


        if choice == '1':
            pool = lower + upper
            level_name = "Easy"
        elif choice == '2':
            pool =  digits + lower + upper
            level_name = "Medium"
        elif choice == '3':
            pool = lower + upper + digits + symbols
            level_name = "Hard"
        else:
            print("Invalid choice, choice between 1/2/3")

        password = ''.join(random.SystemRandom().choice(pool) for _ in range(length))
        
        print(f"\n[{level_name} Password]: {password}")

    except ValueError:
        print("Error: Please enter a valid number for the length.")

if __name__ == "__main__":
    custom_password()
```
